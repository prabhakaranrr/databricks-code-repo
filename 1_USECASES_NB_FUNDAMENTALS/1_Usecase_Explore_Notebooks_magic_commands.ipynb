```python

from pyspark.sql.session import SparkSession
print(spark)
spark1=SparkSession.builder.appName("Spark DataFrames").getOrCreate()
print(spark1)
df1=spark1.read.csv("/Volumes/workspace/default/mobile_metrics/mobile_os_usage.csv",header=True)

df_clean = (df1
    .withColumnRenamed("Mobile Operating System", "mobile_operating_system")
    .withColumnRenamed("Percent of Usage", "percent_of_usage")
)

df_clean.write.format("delta").option("mergeSchema", "true").mode("overwrite").saveAsTable("default.mobile_os_usage")
display(df_clean)
```
